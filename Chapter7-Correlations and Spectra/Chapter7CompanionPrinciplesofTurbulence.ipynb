## 7.1 Two- Point- Correlation


For a velocity field $\mathbf{u}(\mathbf{x},t)$, the (Cartesian) two‑point correlation is
$$
R_{ij}(\mathbf{r}) \;=\; \langle u_i(\mathbf{x},t)\,u_j(\mathbf{x}+\mathbf{r},t)\rangle,
$$
with ensemble/space average $\langle\cdot\rangle$. The **autocorrelation** is $R_{ii}(0)=\langle u_i^2\rangle$. The **normalized** correlation is
$$
\rho_{ij}(\mathbf{r}) \;=\; \frac{R_{ij}(\mathbf{r})}{\sqrt{R_{ii}(0)\,R_{jj}(0)}}.
$$

### Matrix (Tensor) Form
At a fixed separation $\mathbf{r}$, assemble the correlation tensor:
$$
\mathbf{R}(\mathbf{r}) \;=\;
\begin{bmatrix}
R_{11}(\mathbf{r}) & R_{12}(\mathbf{r}) & R_{13}(\mathbf{r})\\
R_{21}(\mathbf{r}) & R_{22}(\mathbf{r}) & R_{23}(\mathbf{r})\\
R_{31}(\mathbf{r}) & R_{32}(\mathbf{r}) & R_{33}(\mathbf{r})
\end{bmatrix}.
$$


# 7.2. Homogeneous/Isotropic Simplifications

Under **homogeneity**, $R_{ij}$ depends only on $\mathbf{r}$, not on $\mathbf{x}$. Under **isotropy**, the tensor structure reduces to two scalar functions, commonly expressed in terms of **longitudinal** and **transverse** correlations.

## A. Longitudinal/Transverse Correlations
Let $\hat{\mathbf{r}}=\mathbf{r}/r$. The **longitudinal** correlation is
$$
R_{LL}(r) \;=\; \langle \, (\mathbf{u}\cdot \hat{\mathbf{r}})(\mathbf{u}'\cdot \hat{\mathbf{r}})\, \rangle,
$$
and the **transverse** correlation is
$$
R_{NN}(r) \;=\; \frac{1}{2}\,\langle \, |\mathbf{u}|^2 + |\mathbf{u}'|^2 - 2(\mathbf{u}\cdot \hat{\mathbf{r}})(\mathbf{u}'\cdot \hat{\mathbf{r}})\, \rangle\Big/2,
$$
evaluated at two points separated by $r$. In 1D demos below we focus on $R_{LL}(r)$ aligned with the $x$–direction.

## B. Integral Length Scale
The **integral length scale** characterizes the correlation extent of turbulence:
$$
L \;=\; \int_{0}^{\infty} \frac{R_{LL}(r)}{R_{LL}(0)} \, dr.
$$
Numerically we truncate the integral where $R_{LL}(r)$ decays to near zero.


## C. Taylor’s Frozen‑Turbulence Hypothesis

With a large mean advection $U$, temporal fluctuations at a fixed probe can be interpreted as spatial structure convected past the probe:
$$
r \;\approx\; U\,\tau,
$$
so that the temporal autocorrelation $R_{uu}(\tau)$ maps to the spatial correlation $R_{uu}(r)$ via $r=U\tau$. This allows estimating spatial scales (e.g., $L$) from time series when direct spatial measurements are unavailable.

## Example NB 7.1. Synthetic 1-D Homogeneous Turbulence Data

We synthesize 1D homogeneous fields by filtering white noise to impose a target correlation length.  it is sufficient to illustrate definitions, algorithms, and scaling.


This example demonstrates how to numerically synthesize a one-dimensional homogeneous random velocity field (u(x)) that mimics the statistical structure of turbulence.  
The field is not a real turbulent flow but reproduces key spatial characteristics such as finite variance and a prescribed correlation length.


In turbulence theory, a *homogeneous field* has statistical properties that are invariant under translation.  
Here we generate such a field synthetically by **filtering white noise** through an exponential kernel, producing a smooth random signal whose fluctuations decay over a characteristic distance called the **correlation length**, \( \ell_c \).

The process can be summarized as:
1. Generate zero-mean, unit-variance **white noise** \( w(x) \).  
   Each sample is independent, representing uncorrelated fluctuations.
2. Define an **exponential kernel**
   $$
   K(r) = e^{-|r|/\ell_c},
   $$
   which introduces correlation between nearby points.
3. Perform **convolution** between the noise and the kernel (implemented efficiently via FFT multiplication):
   $$
   u(x) = \mathcal{F}^{-1}\!\left[\mathcal{F}(w)\,\mathcal{F}(K)\right].
   $$
   This filtering operation gives \( u(x) \) a smooth, correlated structure.
4. Normalize the resulting field so that it has a specified **standard deviation** \( \sigma \), ensuring
   $$
   \langle u^2 \rangle = \sigma^2.
   $$

---

#### Parameters
- `n`: number of grid points in the domain.  
- `dx`: grid spacing (sets the physical scale of \( x \)).  
- `corr_len`: target correlation length \( \ell_c \), controlling how rapidly \( u(x) \) decorrelates with distance.  
- `sigma`: standard deviation of \( u(x) \).

---

#### Physical Interpretation

- A **small** `corr_len` yields rapidly varying fluctuations — resembling high-Reynolds-number turbulence where eddies are small and short-lived.  
- A **large** `corr_len` produces slowly varying structures — analogous to low-frequency, large-scale eddies.  
- The resulting field is **stationary and homogeneous**, meaning that the mean and variance of \( u(x) \) are constant across space.

---

#### Outcome

Running this code produces arrays \( (x, u) \):
- ( x ): spatial coordinate.  
- ( u(x)): the synthetic velocity field, suitable for computing **two-point correlations**, **energy spectra**, and **integral scales** in later notebook sections.


```python
#Example NB 7.1 Synthesize 1D homogeneous 'turbulent' velocity field u(x)
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(7)

def synth_field(n=4096, dx=0.01, corr_len=0.3, sigma=1.0):
    """Generate a 1D homogeneous random field by exponential filtering of white noise.
    The target correlation length is set by corr_len."""
    
    x = np.arange(n)*dx
    # white noise
    w = rng.standard_normal(n)
    # build an exponential kernel ~ exp(-|r|/corr_len)
    r = np.arange(-n//2, n//2)*dx
    kern = np.exp(-np.abs(r)/max(corr_len,1e-6))
    # FFT-based convolution
    W = np.fft.rfft(w)
    K = np.fft.rfft(kern)
    u = np.fft.irfft(W*K, n=n)
    # normalize variance to sigma^2
    u = sigma * (u - np.mean(u)) / (np.std(u) + 1e-12)
    return x, u

# quick demo generation (values reused later)
x_demo, u_demo = synth_field()
plt.figure()
plt.plot(x, u)
plt.xlabel('x'); plt.ylabel('u(x)')
plt.title("Synthetic homogeneous velocity field")
plt.show()

```


## Example NB 7.2 — Construction Two‑Point Correlation on 1-D Homogenous Data
We compute the discrete two‑point correlation
$$
R_{uu}(m) \;=\; \frac{1}{N-m}\sum_{k=0}^{N-m-1} u_k\,u_{k+m},
$$
and its normalized form $\rho_{uu}(m)=R_{uu}(m)/R_{uu}(0)$.


```python
# Example 7.2.
import numpy as np
import matplotlib.pyplot as plt

def two_point_corr(u):
    u = np.asarray(u)
    u = u - np.mean(u)
    n = u.size
    # full autocovariance via FFT for speed
    U = np.fft.rfft(u, n=2*n)
    S = U * np.conjugate(U)
    acov_full = np.fft.irfft(S).real[:n]
    # unbiased estimator: divide by (n - lag)
    lags = np.arange(n)
    acov = acov_full / (n - lags + 1e-12)
    return acov

def integral_length_scale(r, R):
    R0 = R[0] if R[0] != 0 else 1.0
    rho = R / R0
    # truncate where negative/near zero to avoid integrating noisy tails
    mask = rho > 0
    return np.trapz(rho[mask], r[mask])

# Use demo data
x = x_demo; u = u_demo
dx = x[1]-x[0]

R = two_point_corr(u)
lags = np.arange(R.size)
r = lags * dx
rho = R / (R[0] + 1e-12)

L = integral_length_scale(r, R)
print(f"Estimated integral length scale L ≈ {L:.3f} [units of x]")

fig, ax = plt.subplots()
ax.plot(r, rho)
ax.set_xlabel('r'); ax.set_ylabel('rho_uu(r)')
ax.set_title('Normalized Two‑Point Correlation (1D)')
plt.show()
```

## Example NB 7.3 — Temporal Autocorrelation and Taylor’s Frozen Turbulence Hypothesis
We generate a time series $u(t)$ and compute its autocorrelation $R_{uu}(\tau)$. With a chosen mean speed $U$, set $r=U\tau$ to approximate the spatial correlation.
This example demonstrates how to generate a **synthetic time series of velocity fluctuations** and how to relate its **temporal autocorrelation** to the **spatial two-point correlation** using *Taylor’s frozen-turbulence hypothesis*.

---
In turbulence, one often measures a signal ( u(t) ) at a **fixed spatial location** rather than across space.  
To study the spatial structure of turbulence, we assume that turbulent eddies are “frozen” and advected past the sensor by a mean flow ( U ).  
$$\text{This allows us to link time separation }  ( \tau ) \text{and **space separation** } ( r) through:

$$
r \;=\; U \, \tau.
$$

$$\text{Under this assumption, the **temporal autocorrelation**} ( R_{uu}(\tau) ) \text{directly corresponds to the **spatial two-point correlation**} ( R_{uu}(r)).

---

###  Details

1. **Synthetic Time Series**  
   The function `synth_time_series()` creates a *homogeneous random velocity signal* using an **exponential temporal kernel**, producing a process similar to an **Ornstein–Uhlenbeck (OU)** process.  
   - `tau_c` controls the **correlation time**, analogous to a correlation length in spatial fields.  
   - The convolution is performed using the **FFT method**, ensuring efficiency for large arrays.  
   - The output \( u(t) \) has zero mean and unit variance (scaled by `sigma`).

  
Mathematically, the time series represents:
   $$
   u(t) \;=\; \int_{-\infty}^{\infty} w(t')\, e^{-|t-t'|/\tau_c}\, dt',
   $$
   where \( w(t) \) is Gaussian white noise.

2. **Autocovariance Function**  
   The `autocov()` function computes:
   $$
   R_{uu}(\tau) = \frac{1}{T - \tau} \int u(t) \, u(t+\tau)\, dt,
   $$
   using FFT-based convolution for fast computation.

3. **Normalization**  
   The correlation coefficient is obtained as:
   $$
   \rho_{uu}(\tau) = \frac{R_{uu}(\tau)}{R_{uu}(0)},
   $$
   so that \( \rho_{uu}(0) = 1 \) and \( \rho_{uu}(\tau) \to 0 \) as \( \tau \to \infty \).

4. **Taylor’s Hypothesis Mapping**  
   With a mean advection velocity \( U \), the temporal lag \( \tau \) is mapped to spatial separation \( r = U\tau \).  
   This converts the time-domain correlation into a spatial correlation estimate:
   $$
   \rho_{uu}(r) \;\approx\; \rho_{uu}(\tau) \big|_{r = U\tau}.
   $$

---

### Physical Interpretation of Plots

- The **first plot** shows \( \rho_{uu}(\tau) \) vs. \( \tau \), illustrating how correlation decreases with increasing time lag — a measure of the memory of the turbulent signal.
- The **second plot** maps the same data to \( \rho_{uu}(r) \) vs. \( r = U\tau \), showing how spatial correlation decays with increasing distance between points.
- The slope near \( \tau = 0 \) (or \( r = 0 \)) is related to **Taylor’s microscale**, which characterizes the small-scale structure of turbulence.

---



```python
#Example 7.3
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(11)

def synth_time_series(n=4096, dt=0.005, tau_c=0.15, sigma=1.0):
    # OU-like temporal filter via exponential kernel in time
    t = np.arange(n)*dt
    w = rng.standard_normal(n)
    tau = np.arange(-n//2, n//2)*dt
    kern = np.exp(-np.abs(tau)/max(tau_c,1e-6))
    W = np.fft.rfft(w)
    K = np.fft.rfft(kern)
    u = np.fft.irfft(W*K, n=n)
    u = sigma*(u - np.mean(u))/(np.std(u)+1e-12)
    return t, u

def autocov(u):
    u = u - np.mean(u)
    n = u.size
    U = np.fft.rfft(u, n=2*n)
    S = U*np.conjugate(U)
    acov = np.fft.irfft(S).real[:n] / (n - np.arange(n) + 1e-12)
    return acov

# synthesize time series and compute autocorrelation
t, ut = synth_time_series()
dt = t[1]-t[0]
Rtau = autocov(ut)
tau = np.arange(Rtau.size)*dt
rho_tau = Rtau / (Rtau[0] + 1e-12)

# Taylor mapping with chosen U
U = 5.0
r_map = U * tau
rho_r = rho_tau  # same values, different abscissa

fig, ax = plt.subplots()
ax.plot(tau, rho_tau)
ax.set_xlabel('tau'); ax.set_ylabel('rho_uu(tau)')
ax.set_title("Temporal Autocorrelation")
plt.show()

fig, ax = plt.subplots()
ax.plot(r_map, rho_r)
ax.set_xlabel('r = U tau'); ax.set_ylabel('rho_uu(r)')
ax.set_title("Spatial Correlation via Taylor's Hypothesis (mapped)")
plt.show()
```

## Example NB 7.4 — Energy Spectrum from Correlation
For a 1D field $u(x)$, the (one‑sided) energy spectrum $E(k)$ relates to the correlation via Fourier transform. Numerically we compute
$$
E(k) \;\propto\; |\hat{u}(k)|^2,
$$
and compare with the correlation.

This example illustrates how to compute and interpret the **energy spectrum** \(E(k)\) of a one-dimensional velocity field \(u(x)\).  
The energy spectrum represents how the kinetic-energy content of turbulence is distributed among different spatial scales (or equivalently, wavenumbers).

---

### Concept

In homogeneous turbulence, the **two-point correlation function** \(R_{uu}(r)\) and the **energy spectrum** \(E(k)\) are a **Fourier transform pair**:

$$
R_{uu}(r) \;=\; \int_0^{\infty} E(k)\,\cos(k r)\,dk,
\qquad
E(k) \;=\; \frac{2}{\pi} \int_0^{\infty} R_{uu}(r)\,\cos(k r)\,dr.
$$

Thus:
- \(R_{uu}(r)\) describes how velocity fluctuations remain correlated over distance.
- \(E(k)\) shows how energy is partitioned among large-scale (low \(k\)) and small-scale (high \(k\)) motions.

The two descriptions contain the **same statistical information**, just expressed in different domains.

---

### Computational Steps

1. **Remove the Mean and Compute the FFT**  
   The mean velocity is subtracted to isolate fluctuations.  
  $$\text{The discrete Fourier transform of u(x) is:  }  
   \hat{u}(k) = \mathcal{F}[u(x)] \text{ , computed via `np.fft.rfft()` for real signals}$$.

2. **Wavenumber Array**  
   $$\text{The discrete wavenumbers are obtained as: }  
   k = \frac{2\pi m}{L} ,  
   \text{where } L = n\,\Delta x\ \text{ is the domain length}$$.

3. **Energy Spectrum Estimate**  
   The one-sided spectrum is formed from the squared Fourier amplitude:  
   $$
   E(k) \;\propto\; |\hat{u}(k)|^2.
   $$  
   The factor \(2/n\) ensures correct normalization for a single-sided (positive \(k\)) representation.

4. **Comparison with Spatial Correlation**
$$\text{The two-point correlation }  \rho_{uu}(r)  \text{ is plotted alongside E(k).  
   A broad correlation (slowly decaying  } R_{uu}) \text{ corresponds to concentration of energy at **low k** (large eddies),  
   while a short correlation (rapidly decaying } R_{uu} \text{corresponds to more energy at **high k ** (small eddies)}$$.



```python
#Example NB 7.4
import numpy as np
import matplotlib.pyplot as plt

# Use the spatial field u(x)
x = x_demo; u = u_demo
dx = x[1]-x[0]
n = u.size

# FFT and spectrum (one-sided)
Uk = np.fft.rfft(u - np.mean(u))
k = np.fft.rfftfreq(n, d=dx) * 2*np.pi  # radial wavenumber
E = (np.abs(Uk)**2) * (2.0/ n)          # proportional energy density

# Correlation for reference
R = two_point_corr(u)
rho = R/(R[0]+1e-12)
r = np.arange(R.size)*dx

fig, ax = plt.subplots()
ax.plot(k, E)
ax.set_xlabel('k'); ax.set_ylabel('E(k) (arb. units)')
ax.set_title('One‑Dimensional Energy Spectrum')
ax.set_xscale('log'); ax.set_yscale('log')
plt.show()

fig, ax = plt.subplots()
ax.plot(r, rho)
ax.set_xlabel('r'); ax.set_ylabel('rho_uu(r)')
ax.set_title('Correlation (for comparison)')
plt.show()
```

## Example NB 7.5: Interactive Exercise using Homogenous Data

```python
# === Interactive Lab: Homogeneous field, correlation, integral scale, spectrum ===
import numpy as np
import matplotlib.pyplot as plt

# --- helpers (rely only on NumPy) ---
def two_point_corr_fft(u):
    """Unbiased autocovariance via FFT (lags >= 0)."""
    u = np.asarray(u) - np.mean(u)
    n = u.size
    U = np.fft.rfft(u, n=2*n)
    S = U * np.conjugate(U)
    acov = np.fft.irfft(S).real[:n]
    lags = np.arange(n)
    acov = acov / (n - lags + 1e-12)
    return acov

def integral_length_scale(r, R):
    """Integral length from normalized correlation; truncate where rho<=0."""
    if R[0] == 0:
        return 0.0
    rho = R / R[0]
    mask = rho > 0
    if not np.any(mask):
        return 0.0
    return np.trapz(rho[mask], r[mask])

def spectrum_1d(u, dx):
    """One-sided 1D energy spectrum (proportional) from FFT."""
    u = np.asarray(u) - np.mean(u)
    n = u.size
    Uk = np.fft.rfft(u)
    k = np.fft.rfftfreq(n, d=dx) * 2*np.pi  # radial wavenumber
    E = (np.abs(Uk)**2) * (2.0 / n)         # proportional density
    return k, E

# --- try ipywidgets; else fallback to matplotlib sliders ---
try:
    import ipywidgets as W
    from IPython.display import display
    HAS_WIDGETS = True
except Exception:
    HAS_WIDGETS = False

# --- the generator from your earlier cell (referenced) ---
# def synth_field(n=4096, dx=0.01, corr_len=0.3, sigma=1.0): ...

def _run(n=4096, dx=0.01, corr_len=0.3, sigma=1.0, seed=7, show_hist=False, log_spectrum=True):
    # regenerate field on each parameter change
    rng = np.random.default_rng(int(seed))
    # clone of synth_field that uses chosen seed
    def _synth(n, dx, corr_len, sigma):
        x = np.arange(n)*dx
        w = rng.standard_normal(n)
        r = np.arange(-n//2, n//2)*dx
        kern = np.exp(-np.abs(r)/max(corr_len,1e-6))
        W = np.fft.rfft(w)
        K = np.fft.rfft(kern)
        u = np.fft.irfft(W*K, n=n)
        u = sigma * (u - np.mean(u)) / (np.std(u) + 1e-12)
        return x, u

    x, u = _synth(int(n), float(dx), float(corr_len), float(sigma))
    dx_ = x[1] - x[0]

    # correlation + integral scale
    R = two_point_corr_fft(u)
    r = np.arange(R.size) * dx_
    rho = R / (R[0] + 1e-12)
    L = integral_length_scale(r, R)

    # spectrum
    k, E = spectrum_1d(u, dx_)

    # --- plots ---
    plt.figure(figsize=(8,3))
    plt.plot(x, u)
    plt.xlabel('x'); plt.ylabel('u(x)')
    plt.title(f"Synthetic homogeneous field  (n={len(x)}, dx={dx_:.3g}, ℓ_c≈{corr_len})")
    plt.tight_layout()
    plt.show()

    if show_hist:
        plt.figure(figsize=(5,3))
        plt.hist(u, bins=60, density=True)
        plt.xlabel('u'); plt.ylabel('p(u)')
        plt.title("Histogram of u(x)")
        plt.tight_layout()
        plt.show()

    plt.figure(figsize=(6,3.2))
    plt.plot(r, rho)
    plt.xlabel('r'); plt.ylabel(r'$\rho_{uu}(r)$')
    plt.title(f"Normalized two-point correlation;  integral scale L ≈ {L:.3f}")
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6,3.2))
    plt.plot(k, E)
    plt.xlabel('k'); plt.ylabel('E(k) (arb.)')
    plt.title("One-dimensional energy spectrum")
    if log_spectrum:
        plt.xscale('log'); plt.yscale('log')
    plt.grid(True, which='both', alpha=0.3)
    plt.tight_layout()
    plt.show()

if HAS_WIDGETS:
    ui = W.VBox([
        W.HBox([
            W.IntSlider(value=4096, min=512, max=16384, step=512, description='n'),
            W.FloatLogSlider(value=0.01, base=10, min=-3, max=-1, step=0.1, description='dx'),
            W.FloatSlider(value=0.30, min=0.02, max=1.0, step=0.01, description='corr_len'),
        ]),
        W.HBox([
            W.FloatSlider(value=1.0, min=0.1, max=3.0, step=0.05, description='sigma'),
            W.IntSlider(value=7, min=0, max=9999, step=1, description='seed'),
            W.Checkbox(value=False, description='show_hist'),
            W.Checkbox(value=True, description='log_spectrum'),
        ]),
    ])
    out = W.interactive_output(_run, {
        'n': ui.children[0].children[0],
        'dx': ui.children[0].children[1],
        'corr_len': ui.children[0].children[2],
        'sigma': ui.children[1].children[0],
        'seed': ui.children[1].children[1],
        'show_hist': ui.children[1].children[2],
        'log_spectrum': ui.children[1].children[3],
    })
    display(Markdown("**Interactive Lab — Homogeneous Field, Correlation, Integral Scale, Spectrum**"))
    display(ui, out)
else:
    # Fallback: matplotlib sliders
    from matplotlib.widgets import Slider, CheckButtons

    # initial values
    n0, dx0, cl0, s0, seed0 = 4096, 0.01, 0.3, 1.0, 7
    _run(n0, dx0, cl0, s0, seed0, False, True)

    # set up a small slider panel for corr_len and sigma only (portable)
    fig, ax = plt.subplots(figsize=(6,1))
    plt.subplots_adjust(left=0.25, bottom=0.45)
    ax.axis('off')
    ax_cl = plt.axes([0.25, 0.30, 0.65, 0.03])
    ax_sg = plt.axes([0.25, 0.20, 0.65, 0.03])
    ax_dx = plt.axes([0.25, 0.10, 0.65, 0.03])

    s_cl = Slider(ax_cl, 'corr_len', 0.02, 1.0, valinit=cl0, valstep=0.01)
    s_sg = Slider(ax_sg, 'sigma', 0.1, 3.0, valinit=s0, valstep=0.05)
    s_dx = Slider(ax_dx, 'dx', 0.001, 0.1, valinit=dx0, valstep=0.001)

    def _update(_):
        plt.close('all')
        _run(n0, s_dx.val, s_cl.val, s_sg.val, seed0, False, True)

    s_cl.on_changed(_update); s_sg.on_changed(_update); s_dx.on_changed(_update)
    _update(None)
```

# 7.5. Second-Order Longitudinal Structure Function from a Single Trace

**Goal.** In grid-generated (approximately homogeneous, isotropic) turbulence, estimate the **second-order longitudinal structure function** from a **single time series** \(u(t)\) by invoking **Taylor’s frozen-flow hypothesis** \( r = U_\infty\,\tau \).

---

#### 1) Definition

For a velocity component aligned with the separation r (longitudinal),
the **second-order structure function** is
$$
D_{LL}(r) \;=\; \big\langle\,[u(x+r)-u(x)]^{2}\,\big\rangle .
$$

$$\text{If the flow is locally homogeneous/isotropic and advection dominates, a time lag  }
 \tau  \text{at a fixed probe maps to space via Taylor: }
r \;=\; U_\infty\,\tau, \qquad
D_{LL}(r) \;\approx\; \big\langle\,[u(t+\tau)-u(t)]^{2}\,\big\rangle .
$$

$$\text{Equivalently, using the two-point correlation } R_{uu}(r)=\langle u(x)u(x+r)\rangle,
D_{LL}(r) \;=\; 2\,[R_{uu}(0)-R_{uu}(r)] .
$$

---

#### 2) Expected scalings

- **Dissipation range (very small \(r\))**  
  Viscous smoothing yields a **quadratic** scaling:
  $$ D_{LL}(r) \;\propto\; r^{2}. $$
  In Kolmogorov theory, \( D_{LL}(r) \sim (\varepsilon/15\nu)\,r^{2} \).

- **Inertial range (intermediate \(r\))**  
  The Kolmogorov 1941 (K41) prediction is the **\(2/3\)-law**:
  $$ D_{LL}(r) \;\sim\; C_2\,(\varepsilon\,r)^{2/3}, $$
  $$\text{where C_2 is an  } \mathcal{O}(1) \text{constant}$$.

On a log–log plot, these appear as straight lines with slopes **2** and **\(2/3\)**, respectively.

---

#### 3) What the code does

1. **Synthesize a hot-wire-like time series** u(t) whose spectrum is close to k^{-5/3} between
   large-scale and small-scale cutoffs (random phases → realistic fluctuations).
2. Compute
   \( D_{LL}(r) = \langle [u(t+\tau)-u(t)]^2 \rangle \)
   over many lags \( \tau \).
3. Use **Taylor’s hypothesis** with a chosen mean speed \(U_\infty\) to convert \( \tau \mapsto r \).
4. Plot \( D_{LL}(r) \) on log–log axes with **reference guides** proportional to \( r^{2/3} \) (inertial)
   and \( r^{2} \) (dissipation), normalized to intersect the data for easy visual comparison.

---

#### 4) How to read the plots

- The **structure-function curve** \( D_{LL}(r) \) (solid line) should follow the **\(r^{2/3}\)** guide
  over a range of separations (your inertial range).  
- At the **smallest separations**, the curve bends upward toward the **\(r^{2}\)** guide,
  indicating viscous/dissipation effects.
- At **large separations**, \( D_{LL}(r) \) saturates toward \(2\,u'^2\) as the increments decorrelate.

> **Practical tip.** In real data, widen the record length and ensure \(U_\infty\) is large enough
> (high advective ratio) so that Taylor’s hypothesis is appropriate; averages then stabilize and
> the inertial-range slope is easier to see.

## Problem NB 7.7 — Estimating the Integral Length Scale \(L\) and Taylor Microscale \(\lambda\)


### Objective
$$
\text{Given a synthetic homogeneous velocity field } u(x), 
\text{ compute the integral length scale } L 
\text{ and the Taylor microscale } \lambda.
$$

---

### Step 1 — Integral Length Scale

$$
L \;=\; \int_{0}^{\infty} \frac{R_{uu}(r)}{R_{uu}(0)} \, dr,
$$

$$
\text{where } R_{uu}(r) = \langle u(x) \, u(x+r) \rangle 
\text{ is the two-point velocity correlation function.}
$$

$$
\text{Physically, } L \text{ represents the distance over which velocity fluctuations remain correlated. 
It characterizes the size of the energy-containing eddies in the flow.}
$$

---

### Step 2 — Taylor Microscale

$$
\rho(r) \;=\; \frac{R_{uu}(r)}{R_{uu}(0)} 
\;\approx\; 1 - \frac{r^2}{2\lambda^2}, \qquad (r \to 0)
$$

$$
\text{The Taylor microscale } \lambda 
\text{ measures the curvature of } \rho(r) 
\text{ near } r = 0, 
\text{ representing the scale of the smallest coherent motions 
that still retain some correlation.}
$$

$$
\text{It can be estimated by fitting } 
\rho(r) \text{ vs. } r^2 
\text{ near } r=0 \text{ with a line } 
\rho = a + b\,r^2,
\text{ where } b \approx -\frac{1}{2\lambda^2}.
$$

---

### Step 3 — Numerical Procedure

1. $$\text{Generate a synthetic field } u(x) \text{ using exponential filtering of white noise.}$$  
2. $$\text{Compute the correlation function } R_{uu}(r) \text{ via FFT.}$$  
3. $$\text{Normalize to obtain } \rho(r) = R_{uu}(r)/R_{uu}(0).$$  
4. $$\text{Numerically integrate } \rho(r) \text{ up to the first zero crossing to estimate } L.$$
5. $$\text{Fit the first few points of } \rho(r) \text{ vs. } r^2 
   \text{ to estimate } \lambda \text{ using } b = -1/(2\lambda^2).$$

---

### Step 4 — Physical Interpretation

- $$L \text{ increases with the overall correlation length of the field, representing the dominant eddy size.}$$  
- $$\lambda \text{ is typically much smaller than } L \text{ and reflects local smoothness or curvature of } \rho(r).$$  
- $$\text{The ratio } L/\lambda \text{ provides a measure of the turbulent Reynolds number at the small scales.}$$

```python
# === Interactive: Estimate integral length scale L and Taylor microscale λ ===
import numpy as np, matplotlib.pyplot as plt
import ipywidgets as W
from IPython.display import display, Markdown

# ---------- Try ipywidgets; fallback to static run ----------
try:
    import ipywidgets as W
    from IPython.display import display, Markdown
    HAS_WIDGETS = True
except Exception:
    HAS_WIDGETS = False

def synth_field(n=4096, dx=0.01, corr_len=0.3, sigma=1.0, seed=7):
    """Homogeneous 1D random field by exponential filtering of white noise."""
    rng = np.random.default_rng(int(seed))
    x = np.arange(n) * dx
    w = rng.standard_normal(n)
    r = np.arange(-n//2, n//2) * dx
    kern = np.exp(-np.abs(r) / max(corr_len, 1e-9))
    W = np.fft.rfft(w)
    K = np.fft.rfft(kern)
    u = np.fft.irfft(W * K, n=n)
    u = sigma * (u - np.mean(u)) / (np.std(u) + 1e-12)
    return x, u

def two_point_corr_fft(u):
    """Unbiased autocovariance for non-negative lags via FFT."""
    u = np.asarray(u) - np.mean(u)
    n = u.size
    U = np.fft.rfft(u, n=2*n)
    S = U * np.conjugate(U)
    acov = np.fft.irfft(S).real[:n]
    lags = np.arange(n)
    acov = acov / (n - lags + 1e-12)  # unbiased
    return acov

def integral_length_scale(r, R):
    """Integral scale from normalized correlation (truncate where rho<=0)."""
    if R[0] == 0:
        return 0.0
    rho = R / R[0]
    mask = rho > 0
    if not np.any(mask):
        return 0.0
    return np.trapz(rho[mask], r[mask])

def taylor_microscale_from_fit(r, rho, fit_points=8):
    """
    Estimate λ using ρ(r) ≈ 1 − r²/(2λ²).
    Fit ρ = a + b r² near r=0; λ = sqrt(−1/(2b)).
    """
    m = max(3, int(fit_points))
    r2 = (r[:m]**2)
    y  = rho[:m]
    A = np.vstack([np.ones_like(r2), r2]).T
    a, b = np.linalg.lstsq(A, y, rcond=None)[0]
    if b >= 0:  # non-physical; should be negative near 0
        return np.nan, (a, b)
    lam = np.sqrt(-1.0/(2.0*b))
    return lam, (a, b)


def _run(n=4096, dx=0.01, corr_len=0.30, sigma=1.0, seed=7, fit_points=8, show_hist=False):
    # Synthesize field
    x, u = synth_field(int(n), float(dx), float(corr_len), float(sigma), int(seed))
    dx_ = x[1] - x[0]

    # Correlation
    R = two_point_corr_fft(u)
    r = np.arange(R.size) * dx_
    rho = R / (R[0] + 1e-12)

    # Integral scale
    L = integral_length_scale(r, R)

    # Taylor microscale
    lam, (a, b) = taylor_microscale_from_fit(r, rho, fit_points=int(fit_points))

    # --- Plots ---
    # Field
    plt.figure(figsize=(8,3))
    plt.plot(x, u, lw=0.8)
    plt.xlabel('x'); plt.ylabel('u(x)')
    plt.title(f"Synthetic homogeneous field  (n={len(x)}, dx={dx_:.3g}, corr_len={corr_len})")
    plt.tight_layout(); plt.show()

    if show_hist:
        plt.figure(figsize=(5,3))
        plt.hist(u, bins=60, density=True)
        plt.xlabel('u'); plt.ylabel('p(u)')
        plt.title("Histogram of u(x)")
        plt.tight_layout(); plt.show()

    # Correlation + annotations
    plt.figure(figsize=(6.8,3.4))
    plt.plot(r, rho, label=r'$\rho_{uu}(r)$')
    # Small-r quadratic fit overlay: a + b r^2
    r_fit = np.linspace(0, r[int(fit_points*1.2)], 100)
    rho_fit = a + b * r_fit**2
    plt.plot(r_fit, rho_fit, '--', label=f'fit near 0: a={a:.3f}, b={b:.3e}')
    plt.xlabel('r'); plt.ylabel(r'$\rho_{uu}(r)$')
    plt.title(f"Normalized correlation;  L ≈ {L:.4g} ,  λ ≈ {lam:.4g} (units of x)")
    plt.grid(True, alpha=0.3); plt.legend(frameon=False)
    plt.tight_layout(); plt.show()

    # Text report
    print(f"Integral length scale  L ≈ {L:.6g}")
    print(f"Taylor microscale      λ ≈ {lam:.6g}")
    print(f"Fit (ρ ≈ a + b r²) → a={a:.6f}, b={b:.6e}  |  fit_points={int(fit_points)}")

if HAS_WIDGETS:
    ui = W.VBox([
        W.HBox([
            W.IntSlider(value=4096, min=1024, max=16384, step=512, description='n'),
            W.FloatLogSlider(value=0.01, base=10, min=-3, max=-1, step=0.1, description='dx'),
            W.FloatSlider(value=0.30, min=0.02, max=1.5, step=0.01, description='corr_len'),
        ]),
        W.HBox([
            W.FloatSlider(value=1.0, min=0.25, max=3.0, step=0.05, description='sigma'),
            W.IntSlider(value=7, min=0, max=9999, step=1, description='seed'),
            W.IntSlider(value=8, min=3, max=30, step=1, description='fit_points'),
            W.Checkbox(value=False, description='show_hist'),
        ]),
    ])
    out = W.interactive_output(_run, {
        'n': ui.children[0].children[0],
        'dx': ui.children[0].children[1],
        'corr_len': ui.children[0].children[2],
        'sigma': ui.children[1].children[0],
        'seed': ui.children[1].children[1],
        'fit_points': ui.children[1].children[2],
        'show_hist': ui.children[1].children[3],
    })
    display(Markdown("**Interactive Exercise — Estimate \(L\) and \(\lambda\) from a Synthetic Field**"))
    display(ui, out)
else:
    # Static fallback with sensible defaults
    _run(n=4096, dx=0.01, corr_len=0.30, sigma=1.0, seed=7, fit_points=8, show_hist=False)
```
