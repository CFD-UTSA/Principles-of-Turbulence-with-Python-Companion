## Problem 1 — Shear mixing read-offs
Using `Chapter3Companion_ShearMixing.ipynb` or the dataset directly:
- Report `nu_t` at `y=0.5`.
- Report `kappa_t` at `y=0.5` for `Pr_t = 0.7, 0.9, 1.2`.
- Compute the integral of `P_theta(y)` over `y`.


```
import pandas as pd, numpy as np
df = pd.read_csv('Chapter3Companion_ScalarMixing_Shear.csv')
y = df[df['Pr_t']==0.9]['y'].values
nu_t = df[df['Pr_t']==0.9]['nu_t_m2_per_s'].values
i = np.argmin(np.abs(y-0.5))
nu_half = nu_t[i]
k07 = df[df['Pr_t']==0.7]['kappa_t_m2_per_s'].values[i]
k09 = df[df['Pr_t']==0.9]['kappa_t_m2_per_s'].values[i]
k12 = df[df['Pr_t']==1.2]['kappa_t_m2_per_s'].values[i]
Pth = df[df['Pr_t']==0.9]['P_theta_K2_per_s'].values
I = np.trapz(Pth, y)
print('nu_t(0.5)=', nu_half)
print('kappa_t(0.5):', {'0.7':k07,'0.9':k09,'1.2':k12})
print('Integral P_theta dy=', I)
```

## Problem 3 — Sensitivity to turbulent Prandtl number
Using the shear dataset, compute and plot `kappa_t(y)` for `Pr_t=0.7, 0.9, 1.2`. Briefly comment on how `Pr_t` redistributes scalar mixing across the layer.


```
import matplotlib.pyplot as plt, pandas as pd
df = pd.read_csv('/mnt/data/Chapter3Companion_ScalarMixing_Shear.csv')
fig, ax = plt.subplots()
for Pr in [0.7, 0.9, 1.2]:
    d = df[df['Pr_t']==Pr]
    ax.plot(d['y'], d['kappa_t_m2_per_s'], label=f'Pr_t={Pr}')
ax.set_xlabel('y [-]'); ax.set_ylabel('kappa_t [m^2/s]')
ax.set_title('Effect of Turbulent Prandtl Number on kappa_t(y)')
ax.legend(frameon=False); plt.show()
```

## Problem 4 — Buoyancy work vs scalar-variance production
Using the stable dataset, evaluate the ratio `|P_b| / (P_theta + 1e-12)` at `y=0.25, 0.5, 0.75` and discuss where buoyancy is most influential relative to scalar-variance production.


```
import numpy as np, pandas as pd
dfs = pd.read_csv('/mnt/data/Chapter3Companion_ScalarMixing_Buoyancy_Stable.csv')
y = dfs['y'].values
Pb = np.abs(dfs['P_b_m2_per_s3'].values)
Pth = dfs['P_theta_K2_per_s'].values
for y0 in [0.25, 0.5, 0.75]:
    i = np.argmin(np.abs(y - y0))
    print(f'y={y0}: |P_b|/(P_theta+1e-12) =', Pb[i]/(Pth[i]+1e-12))
```
