## 1. Homogeneous Isotropic Turbulence (HIT)
Two datasets represent synthetic HIT decays corresponding to different infrared (large-scale) spectra.

### `HIT_dataset_Saffman.csv`
- **Model:** Saffman infrared spectrum with index $m_0=2$.
- **Invariant:** $I(t)=k\,L^{3}$ is approximately constant.
- **Expected decay exponent:** $n=6/5$.
- **Columns:**
  - `t` (s): time
  - `k` (m²/s²): turbulent kinetic energy
  - `L` (m): energy-containing length scale
  - `epsilon` (m²/s³): dissipation, $C_\varepsilon k^{3/2}/L$
  - `dkdt` (m²/s³): time derivative of $k$
  - `residual` (m²/s³): $dk/dt + \epsilon$ (should be small)

### `HIT_dataset_Loitsyansky.csv`
- **Model:** Loitsyansky infrared spectrum with index $m_0=4$.
- **Invariant:** $I(t)=k\,L^{5}$ is approximately constant.
- **Expected decay exponent:** $n=10/7$.
- **Columns:** same as Saffman dataset.


## 2. Grid-generated Turbulence
Two datasets emulate canonical grid-turbulence power-law decay downstream of a regular grid.

### `Grid_dataset_Saffman.csv`
- **Model:** power-law decay with exponent $n=6/5$ (Saffman).
- **Columns:**
  - `x` (m): downstream location
  - `xi` (–): non-dimensional distance $\xi=(x-x_0)/M$
  - `k` (m²/s²): TKE
  - `epsilon` (m²/s³): dissipation (from $-U_\infty\,dk/dx$)

### `Grid_dataset_Loitsyansky.csv`
- **Model:** power-law decay with exponent $n=10/7$ (Loitsyansky).
- **Columns:** same as Saffman grid dataset.


## 3. Legacy Files and General Notes
- `HIT_dataset.csv` and `Grid_dataset.csv` are earlier, single-regime versions; prefer the explicit Saffman/Loitsyansky variants.
- In all datasets, $k$ is the TKE per unit mass and $\varepsilon$ is the dissipation rate. The length scale $L$ in HIT is an **energy-containing** scale.
- For HIT, the `residual` column should be small, reflecting the model balance $dk/dt = -\varepsilon$.


## 4. Quick Verification Snippets
Use the following cells to quickly load datasets and verify basic properties (budget closure, invariants, and slopes).

### 4.1 HIT: Budget closure and invariant (Saffman)

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/mnt/data/HIT_dataset_Saffman.csv')
t = df['t'].values
k = df['k'].values
L = df['L'].values
eps = df['epsilon'].values
dkdt = df['dkdt'].values
res = df['residual'].values

fig, ax = plt.subplots()
ax.plot(t, -dkdt, label='-dk/dt')
ax.plot(t, eps, label='epsilon')
ax.set_xlabel('t [s]')
ax.set_ylabel('Budget terms [m^2/s^3]')
ax.set_title('Homogeneous Isotropic Turbulence Budget — Saffman (m0=2)')
ax.legend(frameon=False)
plt.show()

I = k * (L**3)
drift = (I.max() - I.min()) / I.mean() * 100.0
print(f'Invariant drift (k L^3): {drift:.2f}%')
fig, ax = plt.subplots()
ax.plot(t, I)
ax.set_xlabel('t [s]')
ax.set_ylabel('Invariant k L^3')
ax.set_title('Saffman invariant check (k L^3)')
plt.show()
```

### 4.3 Grid turbulence: slope checks (Saffman)

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dg = pd.read_csv('/mnt/data/Grid_dataset_Saffman.csv')
xi = dg['xi'].values
k  = dg['k'].values
eps= dg['epsilon'].values
i0 = int(0.4*len(xi))
m_k, b_k = np.polyfit(np.log(xi[i0:]), np.log(k[i0:]), 1)
m_e, b_e = np.polyfit(np.log(xi[i0:]), np.log(eps[i0:]), 1)
print(f'k slope ~ {m_k:.3f} (expected ~ -6/5 = -1.2)')
print(f'epsilon slope ~ {m_e:.3f} (expected ~ -(6/5+1) = -2.2)')
fig, ax = plt.subplots(); ax.loglog(xi, k)
ax.set_xlabel('xi = (x-x0)/M')
ax.set_ylabel('k [m^2/s^2]')
ax.set_title('Grid Turbulence — k vs xi (Saffman)')
plt.show()
fig, ax = plt.subplots(); ax.loglog(xi, eps)
ax.set_xlabel('xi = (x-x0)/M')
ax.set_ylabel('epsilon [m^2/s^3]')
ax.set_title('Grid Turbulence — epsilon vs xi (Saffman)')
plt.show()
```
